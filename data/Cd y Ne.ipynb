```python
import pandas as pd
import matplotlib.pyplot as plt
import math
%pylab inline
```

```python
data = pd.read_csv('ne_paralelo.txt', sep=" ", header=None)
data.columns = ["altura", "pixel"]
data1 = pd.read_csv('ne_b_perpend.txt', sep=" ", header=None)
data1.columns = ["altura", "pixel"]

data2 = pd.read_csv('calibracion.txt', sep=" ", header=None)
data2.columns = ["altura", "pixel"]

data2new = pd.read_csv('calibration.txt', header=None)
data2new.columns = ["lambdita"]

datanew = pd.read_csv('parallel.txt', header=None)
datanew.columns = ["lambdita"]

data1new = pd.read_csv('perpendicular.txt', header=None)
data1new.columns = ["lambdita"]

```

```python
plt.subplot(3,1,1)
plt.plot(data2new.lambdita, data2.pixel, 'g')
plt.rc('text', usetex=True)
plt.ylabel(r'\bf{Intensidad}')
plt.title(r'\bf{Picos de intensidad del Ne}')

plt.subplot(3,1,2)
plt.plot(datanew.lambdita, data.pixel)
plt.rc('text', usetex=True)
plt.ylabel(r'\bf{Intensidad}')

plt.subplot(3,1,3)
plt.plot(data1new.lambdita, data1.pixel, 'r')
plt.rc('text', usetex=True)
plt.ylabel(r'\bf{Intensidad}')
plt.xlabel(r'\bf{Longitud de Onda}')
```
